```python
import pickle
import yellowbrick
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from yellowbrick.classifier import ConfusionMatrix
```

```python
with open("cover_type.pkl", "rb") as f:
    X_treinamento, y_treinamento, X_teste, y_teste = pickle.load(f)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 5, p = 2, metric = "minkowski")
knn_classifier.fit(X_treinamento, y_treinamento)
```

```python
y_pred = knn_classifier.predict(X_teste)
```

```python
y_pred
```

```python
accuracy_score(y_pred, y_teste)
```

```python
cm = confusion_matrix(y_pred, y_teste)
cm
```

```python
print(classification_report(y_pred, y_teste))
```

```python
cm = ConfusionMatrix(knn_classifier)
cm.fit(X_treinamento, y_treinamento)
cm.score(X_teste, y_teste)
```
